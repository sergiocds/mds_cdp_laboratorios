### 1

### 2
Realice una gráfica de barras sobre la población de cada ciudad, ordenela de menor a mayor y responda:

a) ¿Cuál es la segunda ciudad más poblada?<br>
b) ¿Qué posición ocupa Granada en el ranking de las ciudades más pobladas?

```python
# Seleccionando campos de interés del Dataframe original
graph_data=poblacion[['Municipio','Poblacion']]
# Ordenando de forma ascendente el dataframe que se graficará
graph_data=graph_data.sort_values(by=['Poblacion'])
# Reseteando indice del dataframe
graph_data = graph_data.reset_index(drop=True)
# Estableciendo tamaño del gráfico. Se fija un tamaño grande debido a la cantidad de datos.
plt.rcParams["figure.figsize"] = (10,200)
# Creando gráfica de barras horizontales
graph_data.plot.barh(x='Municipio',y='Poblacion')
# Estableciendo tamaño del eje x de acuerdo al valor máximo de poblacion del dataframe
plt.gca().set_xlim([0, graph_data['Poblacion'].max()+100000])
# Inviertiendo eje del gráfico
plt.gca().invert_yaxis()
# Definiendo título del gráfico
plt.title('Población por Municipio')
# Definiendo etiquetas para ejes
plt.xlabel('Población')
plt.ylabel('Municipio')
# Definiendo tamaño de texto para valores en ejes
plt.yticks(fontsize=5)
plt.xticks(fontsize=5)
# Mostrando respuesta a preguntas
# Obteniendo penultimo elemento del dataframe ordenado
segundo_municipio=graph_data.iloc[-2,0]
print(f'La segunda ciudad más poblada es {segundo_municipio}.')
# Obteniendo ranking de granada del dataframe ordenado
granada=(graph_data[(graph_data['Municipio']=='Granada')].index).item()+1
print(f'Granada ocupa la posición {granada} en el ranking de poblaciones de los municipios.')
# Mostrando Gráfico
plt.show()
```

Obtenga la media, mediana, desviación estándar, valor máximo y valor mínimo de la población de los municipios de la provincia de Granada.

### 8
Dígame los nombres de los Municipios más cercano y más lejano a Madrid. Para ello debe calcular la distancia en todos ellos y Madrid. No considere a Madrid en el análisis ya que la distancia sería cero.

```python
# Obteniendo coordenadas de origen (Madrid)
x0=poblacion.loc[poblacion['Municipio']=='Madrid','X'].item()
y0=poblacion.loc[poblacion['Municipio']=='Madrid','Y'].item()
# Calculando distancia para cada municipio. La distancia será la longitud de la hipotenusa del triangulo formado entre las coordenadas X y Y de un municipio determinado y Madrid.
poblacion['DistMadrid']=((((poblacion['X'])-x0)**2)+(((poblacion['Y'])-y0)**2))**(1/2)
poblacion['DistMadrid']
# Obteniendo Municipio más cercano y lejano a Madrid
# Excluyendo Madrid
poblacion.loc[poblacion['Municipio']=='Madrid','DistMadrid']=float('NaN')
# Obteniendo municipio más cercano
cercano=poblacion.iloc[poblacion['DistMadrid'].idxmin(),5]
# Obteniendo municipio más lejano
lejano=poblacion.iloc[poblacion['DistMadrid'].idxmax(),5]
# Restableciendo valor de distancia para Madrid
poblacion.loc[poblacion['Municipio']=='Madrid','DistMadrid']=0
# Imprimiendo resultados
print(f'El municipio más cercano a madrid es {cercano}')
print(f'El municipio más alejado de madrid es {lejano}')
```
